# TruthLens — ISOT Fake News Detection (Wrapper Notebook)

This lightweight notebook documents how the main ISOT fake news experiments
are organised in the TruthLens project.

## Where is the full experiment?

- The primary training + MLflow experiment notebook currently lives at:
  `notebooks/notebooks_01_isot_fake_news_detection_mlflow.ipynb`.

You can either continue working in that notebook, or copy its contents into
this file if you prefer a cleaner name for the capstone submission.

## Expected outputs

The training pipeline should produce at least:

- A scikit-learn Pipeline (TF-IDF + LinearSVC or LogisticRegression).
- A persisted model artifact at `artifacts/best_model.joblib`.
- Logged experiments in MLflow under `./mlruns`.

## Integration with the codebase

- The model artifact is loaded by `ml/analyzer.py` and/or `ml/fake_news_classifier.py`.
- High-level API for inference is exposed via Python functions such as
  `ml.analyzer.analyze_text(text: str) -> dict`.

For detailed integration steps, see `docs/capstone/NOTEBOOK_INTEGRATION.md`.
